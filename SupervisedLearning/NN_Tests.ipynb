```python
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.layers import Embedding
from keras.layers import Convolution1D, GlobalMaxPooling1D, Flatten, ZeroPadding2D
from keras.layers import Convolution2D as Conv2D
```

```python
# NN de DeepMind v1


CNN = Sequential()
CNN.add (Conv2D(
filters = 128,
kernel_size = (5,5),
input_shape = (46,19,19),
border_mode='same',
init='uniform',
activation = 'relu'))
for i in range(11):
    CNN.add (Conv2D(filters = 128,kernel_size = (3,3),
        activation = 'relu', border_mode='same', init='uniform' ))
CNN.add (Conv2D(
filters = 1,
kernel_size = (1,1),
border_mode='same',
init='uniform'))

CNN.add(Flatten())

CNN.add(Activation('softmax'))

CNN.summary()
```

```python
# NN de DeepMind v2


CNN = Sequential()
CNN.add(ZeroPadding2D(input_shape = (46,19,19),padding=(2,2), data_format=None))
CNN.add (Conv2D(
filters = 128,
kernel_size = (5,5),

activation = 'relu'))
for i in range(11):
    CNN.add(ZeroPadding2D(padding=(1,1), data_format=None))
    CNN.add (Conv2D(filters = 128,kernel_size = (3,3),
        activation = 'relu' ))
CNN.add (Conv2D(
filters = 1,
kernel_size = (1,1)))

CNN.add(Flatten())

CNN.add(Activation('softmax'))

CNN.summary()
```
